```python
# Import Dependencies
import pandas as pd
```

```python
jan_2019 = pd.read_csv("201901-citibike-tripdata.csv")
```

```python
jan_2020 = pd.read_csv("202001-citibike-tripdata.csv")
```

```python
# Combining both the dataframes
jan_19_20 = pd.concat([jan_2019, jan_2020], ignore_index=True)
```

```python
jan_19_20
```

```python
# Replacing the number for gender: Gender (Zero=unknown; 1=male; 2=female)
jan_19_20['gender'] = jan_19_20['gender'].replace([0,1,2],['Unknown','Male','Female'])
```

```python
jan_19_20 = jan_19_20.dropna(how='any')
```

```python
jan_19_20["start station id"] = jan_19_20["start station id"].astype(int)
```

```python
jan_19_20["end station id"] = jan_19_20["end station id"].astype(int)
```

```python
jan_19_20
```

```python
jan_19_20.to_csv("citibike_jan_19_20.csv", index=False)
```
